The engineer explains that an engine part seems to be missing from the engine, but nobody can figure out which one. If you can add up all the part numbers in the engine schematic, it should be easy to work out which part is missing.

The engine schematic (your puzzle input) consists of a visual representation of the engine. There are lots of numbers and symbols you don't really understand, but apparently any number adjacent to a symbol, even diagonally, is a "part number" and should be included in your sum. (Periods (.) do not count as a symbol.)

Here is an example engine schematic:

467..114..
...*......
..35..633.
......#...
617*......
.....+.58.
..592.....
......755.
...$.*....
.664.598..

In this schematic, two numbers are not part numbers because they are not adjacent to a symbol: 114 (top right) and 58 (middle right). Every other number is adjacent to a symbol and so is a part number; their sum is 4361.

Of course, the actual engine schematic is much larger. What is the sum of all of the part numbers in the engine schematic?

```python
file_input = open(file="./Inputs/Day3.txt", mode='r')
my_input = file_input.read()
```

```python
def isInsideMatrix(r : int, c : int, rows : int, columns : int) -> bool:
    return r >= 0 and r < rows and c >= 0 and c < columns

def getEdje(c_ini : int, c_fin : int, r : int, rows : int, columns : int) -> list:
    edje = []
    for row in range(r-1, r+2):
        for col in range(c_ini -1, c_fin + 1):
            
            if isInsideMatrix(row, col, rows, columns):
                edje.append([row, col])
    
    for c in range(c_ini, c_fin):
        edje.remove([r, c])
    
    return edje

matrix = []

sum_parts = 0
for line in my_input.split('\n'):
    row = []
    for c in line:
        row.append(c)
    
    matrix.append(row)

for i_row, row in enumerate(matrix):
    i_col = 0;
    
    cols = len(matrix[i_row])
    while i_col < cols:
        digits = []
        
        # Searching for the first digit of the number
        while not matrix[i_row][i_col].isdigit():
            i_col += 1;
            
            if i_col >= cols:
                break            
        
        # Saving the digits
        if i_col < cols:
            while matrix[i_row][i_col].isdigit():
                digits.append(matrix[i_row][i_col])
            
                i_col += 1
            
                if i_col >= cols:
                    break
        
        # Obtaining the edje and checking if the number is a part
        if not digits:
            break
        
        n = len(digits)
        edje = getEdje(i_col-n, i_col, i_row, len(matrix), cols)
        is_part = False
        for cell in edje:
            val_cell = matrix[cell[0]][cell[1]] 
            if (not val_cell.isdigit()) and val_cell != '.':
                is_part = True
                break;
        
        if is_part:
            # Adding the part number
            digits.reverse()
            number = 0
            for d, digit in enumerate(digits):
                number += int(digit) * pow(10, d)
            
            sum_parts += number
        
                
print(sum_parts)
```
